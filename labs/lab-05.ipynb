## 1
สถาบันการเงินใช้ระบบดอกเบี้ยทบต้นในการคำนวณยอดเงินคงเหลือในแต่ละปี ถ้าอัตราดอกเบี้ยคือ $ r\% $ ต่อปี เมื่อ $ P_0 $  เป็นเงินต้น และ $ P_n $ เป็นเงินรวมเมื่อสิ้นปีที่ $ n $ เราจะได้แบบจำลองทางคณิตศาสตร์ดีสครีตในรูปสมการผลต่าง คือ
$$ P_{n+1} = (1 + \frac {r} {100}) P_n $$
โดยมีสูตรบีเนต์ คือ 
$$ P_n = (1 + \frac {r} {100})^n P_0 $$

```python
P_0 = 500_000
r = 1.35
year = 20
```

```python
# Pure Python
# Use the different equation

P = [P_0]
N = list(range(0, year+1))

for n in N[:-1]:
    now_P = P[-1]
    new_P = (1 + r/100) * now_P
    P.append(new_P)

print(N)
print(P)
```

```python
# Pure Python
# Use the Binet formula

P = []
N = list(range(0, year+1))
for n in N:
    new_P = ((1+r/100)**n) * P_0
    P.append(new_P)

print(N)
print(P)
```

```python
import matplotlib.pyplot as plt

plt.plot(N, P, marker='o', markersize=4)
plt.grid()
plt.xlabel('Year: n');
plt.ylabel('Money: $P_n$');
```

```python
# Python with numpy
# Use the Binet Formula

import numpy as np

n = np.arange(0, year+1)
P = ((1+r/100)**n) * P_0

print(n)
print(P)

plt.plot(n, P, marker='o', markersize=4)
plt.grid()
plt.xlabel('Year: n');
plt.ylabel('Money: $P_n$');
```

## 2

จากแบบจำลองข้างต้น หากเรานำอัตราเงินเฟ้อ (Inflation Rate) และ การเพิ่มเงินฝากต่อปีเข้าไปเป็นค่าคงที่ที่เท่ากันในทุก ๆ ต้นปีบัญชีด้วย จะทำให้แบบจำลองเปลี่ยนไปดังนี้

$$ P_{n+1} = (1 + \frac{r}{100} - \frac{f}{100})P_n + d $$

เมื่อ $f$ คือ อัตราเงินเฟ้อในรูปของร้อยละต่อปี และ $d$ คือจำนวนเงินที่ฝากเพิ้มเข้าไปในต้นปีของบัญชี


จากแบบจำลองนี้ จงทำการคำนวณว่าเมื่อผ่านไป 20 ปีมูลค่าของเงินในบัญชีจะเป็นเท่าใด หากกำหนดให้ 

$$ P_0 = 500000, r = 1.35, f = 4.25, \text{ และ } d = 12000 $$
