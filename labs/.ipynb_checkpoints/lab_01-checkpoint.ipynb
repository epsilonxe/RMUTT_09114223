# Lottery Model

อ้างอิงจากตัวอย่าง 2.1 จากเอกสารประกอบการสอน

ชมรมแห่งหนึ่งต้องการหารายได้จากการขายสลาก ผลลัพธ์จะเป็นการเปรียบเทียบระหว่างกำไรจำนวนมากที่อาจเกิดขึ้นสำหรับชมรม กับรางวัลที่น่าดึงดูดใจสำหรับผู้ที่ซื้อสลาก

##  Question
เราต้องขายสลากจำนวนกี่ใบจึงจะได้กำไร 

## Mathematical Question
กำหนดให้ $p, C_{1}, C_{2},..., C, O $ จงหา  $ n$ ที่ทำให้ $P > 0 $ ที่ทำให้ $$ P = np - (C + O) > 0 $$
หรือ
$$ P = np - (\sum_{i=1}^{k} C_i+O) > 0 $$ โดยที่ $k \leq n$


เมื่อตัวแปรแต่ละตัว คือ 

<img src="./lab-01-02.png" width="50%" height="auto">




## Solution

เราจะได้กำไรเมื่อ
$$ n > \dfrac{1}{p}\left( \sum_{i=1}^{k} C_i + O   \right) $$


```python

n = 20208
p = 80
k = 10
c = [1000000, 100000, 10000, 5000, 1000, 500, 100, 10, 5, 1]
O = 500000

P = p*n - (sum(c) + O)

z = (sum(c) + O)/p

print('LOTTERY PRIZES')
print('-' * 20)
for i in range(k):
    print(f'PRIZE {i}: {c[i]} THB')
print('-' * 20)

print(f'Total profit = {P}')
print(f'n = {n}')
print(f'z = {z}')



```
