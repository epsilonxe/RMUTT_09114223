ถ้าในแต่ละวันของการรบ กองกำลังของรัฐบาล 100 คน สามารถสังหารกองกำลังของกลุ่มแบ่งแยกดินแดนได้ 10 คน และ 
กองกำลังของกลุ่มแบ่งแยกดินแดน 100 คน สามารถสังหารกองกำลังของรัฐบาลได้ 12 คน


## 1
จงเสนอแบบจำลองทางคณิตศาสตร์ดีสครีตในรูปสมการผลต่างเพื่อใช้ในการพยากรณ์ผลลัพธ์ของสงครามนี้



กำหนดให้ $A_n$ และ $B_n$ คือ จำนวนของกำลังพลฝ่ายรัฐบาลและฝ่ายแบ่งแยกดินแดน ณ วันที่ $n$ ตามลำดับ

\begin{align}
A_{n+1} &= A_n - bB_n\\
B_{n+1} &= B_n - aA_n
\end{align}

เมื่อ $a$ และ $b$ คือความสามารถของกำลังพลฝ่ายรัฐบาลและฝ่ายแบ่งแยกดินแดนที่สามารถกำจัดฝ่ายตรงข้ามได้ ตามลำดับ



เราสามารถเขียนแบบจำลองข้างต้นในรูปของเมทริกซ์ได้ดังนี้

$$ x_{n+1} = Mx_n$$

เมื่อ

$$ M =  \begin{bmatrix} 1 & -b \\ -a & 1 \end{bmatrix} \text{ และ } x_{n} = \begin{bmatrix} A_{n} \\ B_{n} \end{bmatrix}$$


ทั้งนี้สูตรบีเนต์ในการหาผลเฉลยของระบบสมการผลต่างดังกล่าวคือ


$$ x_{n} = M^{n}x_0$$

## 2

จากสมมติฐานที่กำหนดให้ จงพยากรณ์ผลลัพธ์ของสงครามนี้ว่าฝ่ายใดเป็นผู้ชนะ และฝ่ายที่ชนะต้องใช้เวลากี่วันจึงจะกำจัดกองกำลังฝ่ายตรงข้ามได้ทั้งหมด

```python
import numpy as np
import matplotlib.pyplot as plt


days = 14
x_0 = np.array([10000, 8000])

n = np.arange(0, days)
x = np.zeros((days, 2), dtype=np.float64)
a = 10/100
b = 12/100
M = np.array([[1, -b], [-a, 1]])

for r in range(0, days):
    x[r, :] = np.linalg.matrix_power(M, r) @ x_0

    
    
cat = (n.reshape(days, 1),  x)
sol = np.concatenate(cat, axis=1)

with np.printoptions(precision=3, suppress=True):
    print(sol)
```

```python
A = x[:, 0]
B = x[:, 1]

plt.plot(n, A, label='GOV');
plt.plot(n, B, label='REBEL');
plt.grid();
plt.legend();
plt.xticks(n);
```
