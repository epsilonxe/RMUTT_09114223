```python
import numpy as np
import matplotlib.pyplot as plt

def bacSol(yini, k, t):
    y = yini * np.exp(k*t)
    return y
```

```python
y0 = 10000

t1 = 60
t2 = 240

yt1 = 11550
yt2 = 24645
```

# One parameter: $k$

```python
k = (1/(t1-t2)) * np.log(yt1/yt2)

print(f'k = {k:.5f}')
```

```python
time = np.arange(0, 721, 1)
time
```

```python
sol = bacSol(y0, k, time)
sol
```

```python
plt.figure(figsize=(8,5), dpi=200)
plt.grid()
plt.plot(time, sol)

# ข้อมูลจริง
plt.plot([0, t1, t2], [y0, yt1, yt2], 'o')
```

$$ y(t) = 10000 \cdot \exp(0.00421 \cdot t) $$

```python
print(f'time = {time[-1]}, y({time[-1]}) = {sol[-1]}')
```

# Two parameters: $k_1, k_2$ 

```python
k1 = (1/(0-t1)) * np.log(y0/yt1)
k2 = (1/(t1-t2)) * np.log(yt1/yt2)

print(f'k1 = {k1}')
print(f'k2 = {k2}')

time1 = np.arange(0, t1, 1)
time2 = np.arange(t1, 721, 1)

sol1 = bacSol(y0, k1, time1)
sol2 = bacSol(yt1, k2, time2-t1)

plt.figure(figsize=(8,5), dpi=200)
plt.grid()
plt.plot(time1, sol1, 'blue')
plt.plot(time2, sol2, 'blue')


# ข้อมูลจริง
plt.plot([0, t1, t2], [y0, yt1, yt2], 'x')
```

$$ y(t) = \begin{cases}
10000\cdot\exp\big(0.0024\cdot t \big) & \text{ if } 0 \leq t < 60,\\
11550\cdot\exp\big(0.0042\cdot(t-60)\big) & \text{ if } 60 \leq t ,
\end{cases}$$

```python
print(f'time = {time2[-1]}, y({time2[-1]}) = {sol2[-1]}')
```
