# Evacuation Model แบบจำลองการอพยพผู้ประสบภัย

อ้างอิงจากเอกสารประกอบการสอน บทที่ 2 ตัวอย่างที่ 2.2


$$ n(t^*) = n_0 - \left\lfloor \left(\frac{d}{v}\right)pt^* \right\rfloor = 0$$


$$ t^* =  \left\lceil \frac{n_0 v}{dp} \right\rceil $$

```python
# จำนวนผู้ประสบภัย ณ จุดเริ้มต้น (คน)
n0 = 25

# ระยะทางจากจุดประสบภัยถึงจุดปลอดภัย (เมตร)
d = 850

# ความเร็วในการเคลื่อนย้าย (เมตร/วินาที)
v = 1.42

# ความสามารถในการเคลื่อนย้ายผู้ประสบภัย (คน/วินาที^2)
p = 0.00001

# เวลาที่สูญเสียไปก่อนเริ่มการอพยพ
t0 = 0

# เวลาที่ใช้ในการอพยพทั้งหมด
tstar = find_tstar()
T = t0 + tstar
```

```python
import math

def m(t):
    val = math.floor((d/v)*p*t)
    if val > n0:
        val = n0
    return val

def n(t):
    val = n0 - m(t)
    return val

def find_tstar():
    val = math.ceil(n0 * v / (d * p))
    return val
```

```python
import matplotlib.pyplot as plt


time = 60 * 120
list_t = []
list_m = []
list_n = []
for t in range(time):
    list_t.append(t)
    list_m.append(m(t))
    list_n.append(n(t))
    
plt.figure(figsize=(8,4), dpi=255)
plt.plot(list_t, list_m)
plt.plot(list_t, list_n)
plt.grid(visible=True)
plt.xlabel('Time: $t$ (seconds)')
plt.ylabel('Number of People')
plt.show()

print(f't* = {find_tstar()} seconds')
print(f'Total Evacuation Time: T = {t0} + {tstar} = {T} seconds')
```

```python
list_n0 = []
list_T = []
for n0 in range(0, 101):
    tstar = find_tstar()
    T = t0 + tstar
    print(f'n0 = {n0} --> T = {T} seconds = {T/60:.2f} minutes')
    list_n0.append(n0)
    list_T.append(T)
       
```

```python
plt.figure(figsize=(6,4), dpi=255)
plt.plot(list_n0, list_T)
plt.grid(visible=True)
plt.show()
```
